```python
import os
def calculator():
  def addition(a,b):
    return a+b
  def subtraction(a,b):
    return a-b
  def multiplication(a,b):
    return a*b
  def division(a,b):
    if b==0:
      return "Error can be divided by zero {}/{}".format(a,b)
    return a/b
  def clear_screen():
      os.system('cls' if os.name == 'nt' else 'clear')

  #  menue of calculator
  print (" Simple Calculator : ")
  while True :
      print('1 To choose addition ')
      print('2 To choose subtraction ')
      print('3 To choose multiplication ')
      print('4 To choose division ')
      print('5 To clear  ')
      print('6 To choose exit')
      choose =input("Enter the number 1 to 6\n")
    # addition
      if choose =='1':
        try:
          a=float(input("Enter the number : "))
          b=float(input('Enter the number : '))
          sum=addition(a,b)
          print(f"Addition: {a} + {b} = {addition(a,b)}")
          print(f" Addition of number {a} and {b} is {sum}")
        except ValueError:
          print ('invalid input ')
    #  subtraction  of values
      elif choose=='2':
        try:
            a=float(input("Enter  the number : "))
            b=float(input('Enter the number : '))
            sub=subtraction(a,b)
            print(f"Subtraction: {a} - {b} = {subtraction(a,b)}")
            print(f" Subtraction of number {a} and {b} is {sub}")
        except ValueError :
          print('invalid input')
      # Multiplicatin of values
      elif choose=='3':
        try:
          a=float(input("Enter  the number : "))
          b=float(input('Enter the number : '))
          mul=multiplication(a,b)
          print(f"Multiplication: {a} × {b} = {multiplication(a,b)}")
          print(f" Multiplication of number {a} and {b} is {mul}")
        except ValueError :
          print('invalid input')
      # Division of values
      elif choose =='4':
        try:
            a=float(input('Enter the number : '))
            b=float(input("Enter the number : "))
            div=division(a,b)
            print(f"Division: {a} ÷ {b} = {division(a,b)}")
            print(f" division of number {a} and {b} is {div}")
        except ValueError :
          print('invalid input')
      # clear the screen using import os
      elif choose =='5':
        # print('\n'*4,'clear the calculator')  # not clear it put dives four empty line
        # continue
        clear_screen()
        print('clear the calculator input')
      # exit the program
      elif choose =='6':
        print('exit the calculator , Goodbye ! ')
        break
      else :
       print("Invalid input. Enter a number from 1 to 6.")

```
